### Mutable vs Immutable

Why does Python have tuples *AND* lists? Aren't they the same thing?

Not really... Their *usage* is almost identical. However, *modifying* them is not!

Consider the following two snippets of code...

```python
tuple1 = (1, 2)
tuple2 = tuple1
tuple1 += (5, 6)
print(tuple1 is tuple2)
```

```python
list1 = [1, 2]
list2 = list1
list1 += (5, 6)
print(list1 is list2)
```

What happened? Well... tuples **cannot** be changed after their construction while lists can!

This can be generalized in Python (and in any programming language) to the concept of mutable and immutable objects. **Mutable objects** are objects that *can* be modified after their creation. On the other hand, **immutable objects** are objects that *cannot* be modified after their creation. Therefore, making a variation of an immutable object requires creating a *new* one completely!

This looks like a lot of trouble! Why use immutable objects? Can't mutable objects do everything immutable objects can do and more? Well, yes... but that is not always a good thing! Mutable objects can change over time and that can be very problematic and may cause unexpected behaviors!

For example, in the previous code we modified `list1`, right? Just `list1`?

```python
list2
```

So, yes, we did modify `list2` as well!

**Aliasing** is when more than one variable is bound to the same object. This happens all the time but it is particularly significant with mutable objects. It can, for example, be used to share information between parts of the program. However, it can be very problematic and cause unexpected behaviors when used recklessly.

```python
list3 = [1, 2, [3, 4]]

list4 = list(list3)  # This way you get a new object, not aliasing!
print(list4 is list3)
print(list4[2] is list3[2])

import copy
list5 = copy.deepcopy(list3)  # This way you get a new object, with new internal objects!
print(list5 is list3)
print(list5[2] is list3[2])
```

## Classes

As we have discussed, *everything* is an object in Python. Objects are, as you may know, instantiations of *classes* and, as you may have already guessed, you can indeed define your own classes.

```python
class MyClass:
    
    class_variable = [1, 2, 3]
    
    def __init__(self):
        print('I am the constructor!')
        self.variable = 'I am an object variable'
        self._variable = 'I am a private object variable (by convention xD)'
    
    def my_method(self):
        print('I am a method!')
    
    def my_returning_method(self, something):
        return something
    
    def my_variable_returning_method(self):
        return self._variable
    
    def _my_private_method(self):
        print('I am a private method and I "should" not be called from outside the class')
        
    @classmethod
    def my_class_method(cls):
        print('I am a class method and I implicitly get my class as my first argument:', cls)
        
    @staticmethod
    def my_static_method():
        print('I am a class method and I behave just like any normal method, just call me from my class!')
```

```python
print(MyClass)
my_class_instance = MyClass()
print(my_class_instance)
my_class_instance.my_method()
print(my_class_instance.my_returning_method('I am the something!'))
print(my_class_instance.my_variable_returning_method())
print(my_class_instance.variable)
my_class_instance._my_private_method  # while you can, you "shouldn't" do that
print(my_class_instance._variable)    # again, you can but you shouldn't
print(MyClass.class_variable)
MyClass.my_class_method()             # you can it from my_class_instance as well, but... you shouldn't
MyClass.my_static_method()            # again, you can it from my_class_instance as well, but?
```

In case you are wondering, `self` is the instance of the class that the method is called on. Unlike most programming languages, `self` must be explicit in the methods definition. Why? Because, in Python, `instance.method()` is just a shortcut of writing `ClassName.method(instance)`! This has A LOT of implication that are out of the scope of this notebook!

### Inheritance

A class can inherit from another class. In effect, that gives the *child* class all the functionality of the *parent* class. *Child* classes can *override* some of these functionalities if needed.

```python
class MyOtherClass(MyClass):
    
    def my_method(self):
            print("I am overriding my parent's method")
        
    def my_brand_new_method(self):
        print("I am not inherited")
```

```python
print(MyOtherClass)
my_other_instance = MyOtherClass()
print(my_other_instance)
my_other_instance.my_method()
my_other_instance.my_brand_new_method()
print(my_other_instance.my_returning_method('I am the something!'))
print(my_other_instance.my_variable_returning_method())
print(my_other_instance.variable)
MyOtherClass.my_class_method()
MyOtherClass.my_static_method()
```

### Special Methods

There are some special methods that your class inherits *directly* or *indirectly* from the `object` class which represents the base for all classes in Python. These are methods that your class needs when it interacts with a lot of Python's built-in objects, like containers, and you can override them to fine-tune these interactions.

```python
class MySpecialClass:
    
    def __repr__(self):
        return 'A special representation!'
    
    def __str__(self):
        return 'A special string!'
```

```python
my_special_instance = MySpecialClass()
print(my_special_instance)
print([my_special_instance])
```

Other special functions like `__eq__` and `__hash__` are very important when storing your instances in hashing containers such as `set`s and `dict`s.

In short, you can customize every aspect of your class and its interactions with Python built-ins.

**NOTE:** If you come from the Python 2 world, you may think you need to explicitly inherit from `object` to have these functions ([new-style vs old-style](https://docs.python.org/2/reference/datamodel.html#new-style-and-classic-classes)). However, thankfully, in Python 3 all objects inherit from `object` whether or not you include `(object)` in the class declaration.

## Requirement

Now let's use what we learned so far to *estimate $\pi$*!

Consider a dart board shaped like a circle enclosed by a square and you are throwing darts at it. You are very good and you never land a dart off the square but you are not aiming for anything in particular inside it, so a dart has an equal chance of falling anywhere inside the square!

```python
from matplotlib import pyplot as plt
from matplotlib.patches import Circle, Rectangle

ax = plt.axes()
ax.add_patch(Circle((0,0), 1, fill=False, edgecolor='blue', aa=True, lw=5))
ax.add_patch(Rectangle((-1, -1), 2, 2, fill=False, edgecolor='red', aa=True, lw=5))

ax.set_xlim(-1,1)
ax.set_ylim(-1,1)
ax.axis('equal')
ax.grid(False)
ax.set_axis_off()
```

Then, the chance of a dart to fall inside the circle is the ratio of the circle area to the square area. $$p(circle) = \dfrac{circle\ area}{square\ area} = \dfrac{\pi r^2}{(2r)^2} = \dfrac{\pi}{4}$$ This means that $$\pi = 4 * p(circle)$$

Now, the remaining question is, how do we estimate the probability? We can do that by simulating random thrown darts, counting the one ones that land inside the circle and divide that by the total number of thrown darts!

$$p(circle) = \dfrac{number\ of\ darts\ inside\ the\ circle}{total\ number\ of\ darts}$$

That way of estimating probabilites is called *Monte Carlo* simulation.

```python
from matplotlib import pyplot as plt
from matplotlib.patches import Circle, Rectangle

ax = plt.axes()
ax.add_patch(Circle((0,0), 1, fill=False, edgecolor='blue', aa=True, lw=5))
ax.add_patch(Rectangle((-1, -1), 2, 2, fill=False, edgecolor='red', aa=True, lw=5))

x = [random.uniform(-1, 1) for _ in range(1000)]
y = [random.uniform(-1, 1) for _ in range(1000)]
ax.scatter(x, y, s=5)

ax.set_xlim(-1,1)
ax.set_ylim(-1,1)
ax.axis('equal')
ax.grid(False)
ax.set_axis_off()
```
