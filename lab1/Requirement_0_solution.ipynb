```python
import random

from matplotlib import pyplot as plt
from matplotlib.patches import Circle, Rectangle

ax = plt.axes()
ax.add_patch(Circle((0,0), 1, fill=False, edgecolor='blue', aa=True, lw=5))
ax.add_patch(Rectangle((-1, -1), 2, 2, fill=False, edgecolor='red', aa=True, lw=5))
total_number = 1000
x = [random.uniform(-1, 1) for _ in range(total_number)]
y = [random.uniform(-1, 1) for _ in range(total_number)]
ax.scatter(x, y, s=5)

ax.set_xlim(-1,1)
ax.set_ylim(-1,1)
ax.axis('equal')
ax.grid(False)
ax.set_axis_off()
```

```python
q = 0  # number of points in the circles
for i in range(total_number) :
    if (x[i]**2 + y[i]**2)<1:
        q +=1
print(q)
probability_circle = q  / total_number
print(probability_circle)
pi = probability_circle * 4
print(pi)
```
